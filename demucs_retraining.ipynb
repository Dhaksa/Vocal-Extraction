```python
import torch
import sys
from demucs.train import main  # Import main function correctly
import argparse

def retrain_demucs():
    parser = argparse.ArgumentParser(description="Retrain Demucs for Vocal/Background Music Separation")
    parser.add_argument('--root', type=str, default='/content/drive/MyDrive/New/Demucs_Training_Dataset', help='Root directory of dataset')
    parser.add_argument('--epochs', type=int, default=100, help='Number of epochs')
    parser.add_argument('--batch_size', type=int, default=16, help='Batch size')
    parser.add_argument('--learning_rate', type=float, default=3e-4, help='Learning rate')
    parser.add_argument('--model_name', type=str, default='retrained_demucs', help='Name of trained model')
    args = parser.parse_args()

    # Training command with correct argument structure
    train_args = [
        "train",
        f"--root={args.root}",
        f"--epochs={args.epochs}",
        f"--batch_size={args.batch_size}",
        f"--lr={args.learning_rate}",  # Corrected argument
        f"--model={args.model_name}",  # Corrected argument
        "--sources=vocals,background_music",  # Ensuring sources are formatted correctly
        "--workers=4",
        "--device=cuda" if torch.cuda.is_available() else "--device=cpu",
    ]

    sys.argv = [sys.argv[0]] + train_args  # Ensure proper argument passing
    main()

if __name__ == '__main__':
    retrain_demucs()
```

```python
import torch
import torchaudio
import argparse
import os
import numpy as np
from tqdm import tqdm
from torch.utils.data import DataLoader, Dataset
from demucs import Demucs

# Custom dataset for loading audio data
class AudioDataset(Dataset):
    def __init__(self, root_dir):
        self.root_dir = root_dir
        self.files = [f for f in os.listdir(root_dir) if f.endswith('.wav')]  # Load .wav files

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        file_path = os.path.join(self.root_dir, self.files[idx])
        waveform, sample_rate = torchaudio.load(file_path)  # Load audio
        return waveform  # Return audio tensor

# Training function
def train_demucs(model, dataloader, optimizer, epochs, device):
    model.train()
    loss_fn = torch.nn.MSELoss()  # Mean Squared Error for separation loss

    for epoch in range(epochs):
        running_loss = 0.0
        for inputs in tqdm(dataloader, desc=f"Epoch {epoch+1}/{epochs}"):
            inputs = inputs.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)  # Pass through model
            loss = loss_fn(outputs, inputs)  # Compute loss
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        print(f"Epoch {epoch+1}/{epochs} - Loss: {running_loss / len(dataloader)}")

# Main training function
def retrain_demucs(root_dir, epochs, batch_size, learning_rate, model_name):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"Training on {device}")

    # Load dataset
    dataset = AudioDataset(root_dir)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

    # Load Demucs model
    model = Demucs()  # Initialize Demucs model
    model.to(device)

    # Define optimizer
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    # Train the model
    train_demucs(model, dataloader, optimizer, epochs, device)

    # Save trained model
    save_path = f"{model_name}.pth"
    torch.save(model.state_dict(), save_path)
    print(f"Training complete. Model saved as {save_path}")

# Argument parser
if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Train Demucs without Dora/Hydra")
    parser.add_argument('--root', type=str, required=True, help='Root directory of dataset')
    parser.add_argument('--epochs', type=int, default=100, help='Number of epochs')
    parser.add_argument('--batch_size', type=int, default=16, help='Batch size')
    parser.add_argument('--learning_rate', type=float, default=3e-4, help='Learning rate')
    parser.add_argument('--model_name', type=str, default='retrained_demucs', help='Name of trained model')

    args = parser.parse_args()
    retrain_demucs(args.root, args.epochs, args.batch_size, args.learning_rate, args.model_name)

```
